```python
TASK 1 : DATA CLEANING AND PREPROCESSING
DATASET : CUSTOMERS PERSONALITY ANALYSIS
```

```python
import pandas as pd
```

```python

df = pd.read_csv("marketing_campaign.csv", sep="\t")
df.info()
```

```python
df.head()
```

```python
df['Income'] = df['Income'].fillna(df['Income'].median())
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df['Education'] = df['Education'].str.lower().str.strip()
df['Marital_Status'] = df['Marital_Status'].str.lower().str.strip()
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'together': 'married', 'alone': 'single', 'absurd': 'other', 'widow': 'other', 'divorced': 'other'
})
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
print(df.dtypes)
```

```python
df.to_csv("cleaned_marketing_campaign.csv", index=False)
```
